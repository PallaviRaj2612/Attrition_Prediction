```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# columns to encode
encode_cols = ['TravelProfile','Department','EducationField','Gender','Designation','MaritalStatus']

# instance of one-hot encoder
encoder = OneHotEncoder(drop='first', sparse_output=False)

# fit-transform
encoded_arr = encoder.fit_transform(train_drop[encode_cols])

# convert array to dataframe
encoded_df = pd.DataFrame(encoded_arr, columns=encoder.get_feature_names_out(encode_cols))

# merge with main dataframe
train_df = pd.concat([train_drop.drop(columns=encode_cols), encoded_df], axis=1)

train_df.head(20)
```

Droping column

```python
train_df.drop('EmployeeID', axis=1, inplace=True)
```

```python
train_df.head()
```

```python
# Separate numerical and categorical columns
num_cols = train_df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = train_df.select_dtypes(include=['object', 'category']).columns

# Fill numerical NaNs with median
train_df[num_cols] = train_df[num_cols].fillna(train_df[num_cols].median())

# Fill categorical NaNs with mode
for col in cat_cols:
    train_df[col] = train_df[col].fillna(train_df[col].mode()[0])
```

Train and Test split

```python
#import the sklearn library
import sklearn

# import the train_test_split module from sklearn 
from sklearn.model_selection import train_test_split
```

```python
# select the target variable
Y = train_df['Attrition']

# select all the independent variables
# by dropping the target variable
X = train_df.drop(['Attrition'], axis = 1)
```

```python
# let us now split the dataset into train & test
# test_size: the proportion of data to be included in the testing set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state=100)

# print the shape of 'x_train'
print("X_train ",X_train.shape)

# print the shape of 'x_test'
print("X_test ",X_test.shape)

# print the shape of 'y_train'
print("Y_train ",Y_train.shape)

# print the shape of 'y_test'
print("Y_test ",Y_test.shape)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

log_reg = LogisticRegression(max_iter=1000, class_weight='balanced')
log_reg.fit(X_train, Y_train)

y_pred_lr = log_reg.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, y_pred_lr))
print(classification_report(Y_test, y_pred_lr))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
log_reg = LogisticRegression(max_iter=2000, class_weight='balanced')
log_reg.fit(X_train_scaled, Y_train)
```

```python
y_pred_lr = log_reg.predict(X_test_scaled)

print("Accuracy:", accuracy_score(Y_test, y_pred_lr))
print(classification_report(Y_test, y_pred_lr))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

rf = RandomForestClassifier(
    n_estimators=300,              # number of trees
    max_depth=12,                 # limit depth to avoid overfitting
    min_samples_split=4,
    min_samples_leaf=2,
    class_weight='balanced',      # handles imbalance
    random_state=42
)

# Train the model
rf.fit(X_train_scaled, Y_train)

# Predictions
y_pred_rf = rf.predict(X_test_scaled)

# Evaluation
print("Accuracy:", accuracy_score(Y_test, y_pred_rf))
print(classification_report(Y_test, y_pred_rf))

```

```python
Precision 93%

Out of all people predicted as â€œAttrition = Yesâ€, 93% are truly attrition cases.
That means very few false positives â€” excellent

Recall 84%

Here my model is catching 84% of all actual attrition cases.
This is very strong for real-world HR datasets where recall is usually 50â€“70%.
```

```python
Using probability threshold tuning to increase Recall
```

```python
y_prob = rf.predict_proba(X_test_scaled)[:,1]
y_pred_0_45 = (y_prob > 0.45).astype(int)   # lower threshold increases recall
print(classification_report(Y_test, y_pred_0_45))
```

xgboost Model

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report

xgb = XGBClassifier(
    n_estimators=500,
    learning_rate=0.03,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric='logloss',
    scale_pos_weight = (len(Y_train[Y_train==0]) / len(Y_train[Y_train==1])),  # handles imbalance
    random_state=42
)

xgb.fit(X_train_scaled, Y_train)

y_pred_xgb = xgb.predict(X_test_scaled)

print(classification_report(Y_test, y_pred_xgb))
```

```python
print("Accuracy:", accuracy_score(Y_test, y_pred_xgb))
```
